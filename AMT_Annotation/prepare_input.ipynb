```python
random.seed() 
```

```python


effective_question_per_HIT = 10
control_question_per_HIT = 2
```

```python
# The line number start with 0
line_start = 3000   # The boundary included
line_end = 3500   # The boundary excluded

assert (line_end-line_start) % effective_question_per_HIT == 0, "The total number is not a multiple of #question/HIT."
HIT_number = int((line_end-line_start) / effective_question_per_HIT)

#filename_outputs = f"/home/wenjian/Internship/data/AmazonMTurk/inputs/input_{file_prefix}_{line_start}-{line_end-1}-ten.csv"
filename_outputs = f"/media/dongwenjian/SSDBACKUP/newly_added/new_AMT/data/inputs/input_{file_prefix}_{line_start}-{line_end-1}-ten.csv"

print("file_prefix:", file_prefix)
```

```python
filename_outputs
```

```python
df = pd.DataFrame()
df.insert(loc=len(df.columns), column='LineStart', value=[], allow_duplicates=False)
df.insert(loc=len(df.columns), column='LineEnd', value=[], allow_duplicates=False)
df.insert(loc=len(df.columns), column='EffectiveQuestionPerHIT', value=[], allow_duplicates=False)
df.insert(loc=len(df.columns), column='ControlQuestionPerHIT', value=[], allow_duplicates=False)
for i in range(effective_question_per_HIT):
    df.insert(loc=len(df.columns), column='En'+str(i), value=[], allow_duplicates=False)
    df.insert(loc=len(df.columns), column='Fr'+str(i), value=[], allow_duplicates=False)
    df.insert(loc=len(df.columns), column='ImageName'+str(i), value=[], allow_duplicates=False)
for i in range(control_question_per_HIT):
    df.insert(loc=len(df.columns), column='control_En'+str(i), value=[], allow_duplicates=False)
    df.insert(loc=len(df.columns), column='control_Fr'+str(i), value=[], allow_duplicates=False)
    df.insert(loc=len(df.columns), column='control_ImageName'+str(i), value=[], allow_duplicates=False)
```

```python
with open(filename_sample_list, 'r') as con_f:
    sample_list = con_f.readlines()
for i in range(len(sample_list)):
    sample_list[i] = int(sample_list[i][:-1])  # To remove the last '\n'
```

```python
sample_array = np.array(sample_list)
```

```python
sample_array
```

```python
len(sample_array)
```

```python
# Example of one row
# file_prefix**line_number**sentence_flag**original_sentence



line_number = 0  # Attention: this number is for effective answers only. Don't use or modify it when working on control questions

with open(filename_ens_effective, 'r') as f_ens:
    with open(filename_frs_effective, 'r') as f_frs:
        with open(filename_imagenames, 'r') as f_imgs:
            # Jump to the starting line
            for i in range(line_start):
                f_ens.readline()
                f_frs.readline()
                f_imgs.readline()
                line_number += 1
            
            # Load rows for each HIT
            for i in range(HIT_number):
                elt_list = []
                elt_list.append(line_start+i*effective_question_per_HIT)
                elt_list.append(line_start+(i+1)*effective_question_per_HIT)
                elt_list.append(effective_question_per_HIT)
                elt_list.append(control_question_per_HIT)

                # Add effective questions
                for j in range(effective_question_per_HIT): 
                    sentence_flag = "e" # e --> effective
                    elt_list.append(file_prefix+"**"+str(line_number)+"**"+sentence_flag+"**"+(f_ens.readline()[:-1].replace('"', '\\"')))  # [:-1] is to remove the last character, which is a '\n'; quote mark is a special character in Javascript, so excape it.
                    elt_list.append(file_prefix+"**"+str(line_number)+"**"+sentence_flag+"**"+(f_frs.readline()[:-1])) # ** is the delimiter to separate file prefix, sentence flag, line number and the sentence
                    elt_list.append(file_prefix+"**"+str(line_number)+"**"+sentence_flag+"**"+(f_imgs.readline()[:-1]))
                    line_number += 1

                # Add control questions
                for j in range(control_question_per_HIT):
                    sentence_flag = "c"  # c --> control
                    ran = random.randint(0,len(sample_array)-1) # note that random.randint(a,b) returns a random integer N such that a <= N <= b.
                    control_line_number = sample_array[ran]
                    elt_list.append(control_file_prefix+"**"+str(control_line_number)+"**"+sentence_flag+"**"+(list_control_ens[control_line_number][:-1]))  # [:-1] is to remove the last character, which is a '\n'
                    elt_list.append(control_file_prefix+"**"+str(control_line_number)+"**"+sentence_flag+"**"+(list_control_frs[control_line_number][:-1])) # ** is the delimiter to separate file prefix, sentence flag, line number and the sentence
                    elt_list.append(control_file_prefix+"**"+str(control_line_number)+"**"+sentence_flag+"**"+(list_control_imgs[control_line_number][:-1]))

                df.loc[i] = elt_list
```
