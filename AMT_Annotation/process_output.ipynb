## Accuracy Check

```python
class word_object_pair:
    def __init__(self, w, obj_id):
        self.word = w
        self.object_id = obj_id  # For ordinary word without any object, this value is -1
    
    def __str__(self):
        return self.word + ':' + str(self.object_id)
```

```python
def build_object_info_for_each_word(annotated_sentence):
    """
    Argument:
        annotated_sentence: string
        
    """
    list_word_object_pair = []
    
    # First remove the prefix
    annotated_sentence = annotated_sentence.split('**')[3].strip()
    
    token_list_by_space = annotated_sentence.split(" ")
    
    #print(token_list_by_space)
    
    object_state = -1  # -1 is the flag for non-object words
    
    for token in token_list_by_space:
        
        if token[0] == '[':    # example: [/EN#1745/other
            # Extract the number in it
            s = token[re.search("#", token).end():]  # Remove the chars before number chars
            s = s[:re.search("/", s).start()]  # Remove the chars after number chars
            object_state = int(s)
        elif token[-1] == ']':    # base-ball]
            list_word_object_pair.append(word_object_pair(token[:-1], object_state))
            object_state = -1
        else:
            list_word_object_pair.append(word_object_pair(token, object_state))
    return list_word_object_pair
    
```

```python
def accuracy(annotated_sentence_to_evaluate, annotated_sentence_reference):
    list_to_evaluate = build_object_info_for_each_word(annotated_sentence_to_evaluate)
    list_reference = build_object_info_for_each_word(annotated_sentence_reference)
    
    match_count = 0
    for i in range(len(list_to_evaluate)):
        assert list_to_evaluate[i].word == list_reference[i].word, "unmatched word"
        if list_to_evaluate[i].object_id == list_reference[i].object_id:
            #print(list_to_evaluate[i].word)
            match_count += 1
    #print(match_count)
    return match_count/len(list_to_evaluate)
```

## Put all the tests above together

```python
csv_df = pd.read_csv(csv_filepath)

effective_questions_per_HIT = int(csv_df['Input.EffectiveQuestionPerHIT'][0])

# Load standard answers
with open(standard_answer_filepath, "r") as f:
    standard_answers = f.readlines()
    
with open(sentence_outputpath, 'w') as f:
    
    for HIT_index in range(len(csv_df.index)):
        json_str = csv_df.iloc[HIT_index]["Answer.taskAnswers"]
        json_str = json_str[1:-1] # To remove the square bracket
        descriptions_unsorted = json.loads(json_str) # The result is a dictionary
        
        descriptions_effective = []
        descriptions_control = []
        
        # Split the effective sentences and the control sentences into two lists
        for sentence in descriptions_unsorted.values():
            if sentence.split('**')[2] == 'e':
                descriptions_effective.append(sentence)
            elif sentence.split('**')[2] == 'c':
                descriptions_control.append(sentence)
            else:
                raise Exception("sentence flag error")
        
        # Sort the effective sentences list by the original line number
        descriptions_effective_sorted = sorted(descriptions_effective, key=get_key)
        
        # Evaluate accuracy on control sentences
        list_annotated_sentence_to_evaluate = descriptions_control.copy()
        list_annotated_sentence_reference = []
        for sentence_to_evaluate in list_annotated_sentence_to_evaluate:
            assert sentence_to_evaluate.split("**")[2] == 'c'
            line_position = int(sentence_to_evaluate.split("**")[1])
            # Here we no longer assume that all lines in standard answers are consecutive and beginning by 0
            for sentence_standard in standard_answers:
                if int(sentence_standard.split("**")[1]) == line_position:
                    list_annotated_sentence_reference.append(sentence_standard[:-1])  # [:-1] is to remove the last \n

        # Print accurarcies
        accuracies = []
        for i in range(len(list_annotated_sentence_to_evaluate)):
            accuracies.append(accuracy(list_annotated_sentence_to_evaluate[i], list_annotated_sentence_reference[i]))
        average_accuracy = np.mean(np.array(accuracies))
        print("HIT {}:".format(HIT_index), accuracies, average_accuracy)
        
        
        # If test passed, write answers of effective sentences to file
        hit_line_start = csv_df.iloc[HIT_index]['Input.LineStart']
        hit_line_end = csv_df.iloc[HIT_index]['Input.LineEnd']
        if average_accuracy > accuracy_seuil: 
            for sentence in descriptions_effective_sorted:
                f.write(sentence+"\n")
            print(f"Parsed result written to file. Question range: {hit_line_start}-{hit_line_end}")
        else:
            print(f"#### Accuracy failed on control questions. Not written to file. Question range: {hit_line_start}-{hit_line_end} ####")
```
