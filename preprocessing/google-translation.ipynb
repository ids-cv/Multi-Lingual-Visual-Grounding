```python
# For training
for outer_i in range(9):
    begin = outer_i*50000
    end = (outer_i+1)*50000
    with open(f"for-google-translation/queries_{begin}-{end}.txt", 'w') as f:
        try:
            for i in range(begin, end):
                f.write(f"{i}\t" + df_triple.iloc[i]["entity_content"] + "\n")
        except IndexError:
            print(IndexError)
            pass
    print(outer_i, "finished.")
```

```python
df_triple = df_triple.rename(columns = {"entity_content":"entity_content_en"})
```

```python
df_triple
```

```python
df_triple["entity_content"] = ''
```

```python
# For training
for outer_i in range(9):
    outer_i = 0
    begin = outer_i*50000
    end = (outer_i+1)*50000
    with open(f"for-google-translation/fr/queries_{begin}-{end}.fr", 'r') as f:
        try:
            for i in range(begin, end):
                l = f.readline()
                if len(l.split('\t')) == 1:
                    fr_words = l.split(' ')[1:]
                    fr = " ".join(fr_words)
                else:
                    fr = l.split('\t')[1]
                df_triple.at[i, "entity_content"] = fr[:-1]  # [:-1] is to remove '\n'
        except IndexError as e:
            print(IndexError, i)
            raise e
            
    print(outer_i, "finished.")
    break
```

```python
# For validation
begin = 0
end = 14526
with open("for-google-translation/fr/val_queries.fr", 'r') as f:

    for i in range(begin, end):
        l = f.readline()
        if len(l.split('\t')) == 1:
            fr_words = l.split(' ')[1:]
            fr = " ".join(fr_words)
        else:
            fr = l.split('\t')[1]
        df_triple.at[i, "entity_content_fr"] = fr[:-1]  # [:-1] is to remove '\n'
```
