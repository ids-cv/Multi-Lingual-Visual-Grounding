## Running through all the files and getting the data we need: entities mentions

```python
#phase = "val"
#phase = "train"
phase = "test_2016_flickr"

partition_file = "/home/wenjian/Internship/data/flickr30kentities/Multi30k/task1/image_splits/" + phase + ".txt"
output_path = f"/home/wenjian/Internship/data/flickr30kentities/queries_extracted/{phase}_queries.csv"

imageID_list = []
with open(partition_file, 'r') as f:
    l = f.readline()
    while l :
        imageID_list.append(l.split('.')[0])
        l = f.readline()

#Defining regex structure to parse information 
regex = r"\[(.*?)\]"

#Data directory
#directory = 'vgp/flickr30k_entities/annotations/testing_algo'
directory = './annotations/Sentences'
nbr_sentences_file = len(imageID_list)
counter = 0
#Create dictionary to append data and further transform in Dataframe 
dict_df = {'imgId_entityId_entClass':[], 'imgId':[], 'entityId':[], 'entity_content':[], 'entClass':[]}
```

```python


## Iterating over all files flickr30_entities 
for imgId in imageID_list:
    filename = imgId + ".txt"
    f = open(os.path.join(directory,filename))
    if not counter%100: 
        print(str(counter)+'/' + str(nbr_sentences_file))
    counter+=1
    
    #Get next line
    line = f.readline()
    # While the document still have lines, we will continue to read it
    while line:
        #print(line)

        #Defining dataframe variables 
        entityId = ''
        entity_content = ''
        imgId_entityId_entClass = ''
        entClass = ''

        ### Applying regex in order to find the wanted structures 
        test_str = (line)

        ##Get matches 
        matches = re.finditer(regex, test_str, re.MULTILINE)

        ## Iterate over matches and save information
        for matchNum, match in enumerate(matches, start=1):
            for groupNum in range(0, len(match.groups())):

                groupNum = groupNum + 1
                # get entityId by first splitting the '/EN#255542/people A bearded man' in '/' and later splitting in the '#'. We then get the second memeber of the list
                entityId = match.group(groupNum).split('/')[1].split('#')[1]
                #We split the words after the last '/' in order to catch the class and the content 
                words = match.group(groupNum).split('/')[2].split(' ')
                entClass = words[0]
                #We join all the words minus the class word to find our entity content 
                entity_content = ' '.join(words[1:]) 
                #Creating an unique id for the entry
                imgId_entityId_entClass = imgId +'_' +str(entityId)+ '_' + str(entClass)

                '''
                print(entClass)
                print(entityId)
                print(imgId)
                print(imgId_entityId_entClass)
                print(entity_content)
                '''
                #Adding to dict to further add to the df
                dict_df['imgId_entityId_entClass'].append(imgId_entityId_entClass)
                dict_df['imgId'].append(imgId)
                dict_df['entityId'].append(entityId)
                dict_df['entity_content'].append(entity_content)
                dict_df['entClass'].append(entClass)



        #next iteration
        line = f.readline()


    #print('----------------File ended----------------')
        

```

## Exclude entries with no bounding box

```python
import xml.etree.ElementTree as ET
xml_directory = "/home/wenjian/Internship/data/flickr30kentities/annotations/Annotations/"
def have_bounding_box(image_id, object_id): # image_id and object_id are expected to be string
    xml_tree = ET.parse(xml_directory + image_id + '.xml')
    root = xml_tree.getroot()
    for obj in root.findall('object'):
        for name in obj:
            if name.text == object_id:
                if len(obj.findall('bndbox')) >= 1:
                    return True
    return False
```

```python
have_bounding_box("2679015215", "65727")  # For testing
```

```python
def have_bounding_box_for_dataframe(imgId_entityId_entClass):
    image_id = imgId_entityId_entClass.split("_")[0]
    object_id = imgId_entityId_entClass.split("_")[1]
    return have_bounding_box(image_id, object_id)
```

```python
df1 = df[df['imgId_entityId_entClass'].apply(lambda x: have_bounding_box_for_dataframe(x))]
```
